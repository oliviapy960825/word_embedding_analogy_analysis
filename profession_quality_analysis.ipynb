```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_excel("bias_score_updated.xlsx",sheet_name='profession_analogy')
df
```

```python
original=df[['original','original_debias_score_peiyu','original_make_sense_score_peiyu','original_debias_score_luming','original_make_sense_score_luming']]
original['original_debias_score']=(original['original_debias_score_peiyu']+original['original_debias_score_luming'])/2
original['original_make_sense_score']=(original['original_make_sense_score_peiyu']+original['original_make_sense_score_luming'])/2
original=original[['original_debias_score','original_make_sense_score']]
original
```

```python

zero_count=len(original[original['original_make_sense_score']==0])
print(zero_count)

unbiased_count=len(original[original['original_debias_score']==3])
  
print(unbiased_count)


print("avg make sense score is: "+str(round(sum(original.original_make_sense_score)/len(original.original_make_sense_score),2)))

bias_score=original.original_debias_score
bias_score=bias_score[bias_score!=0]
print(bias_score)
print("avg bias score is: "+str(round(sum(bias_score)/len(bias_score),2)))
```
