```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_excel("bias_score_updated.xlsx",sheet_name='female_male_analogy')
df
```

```python
original=df.iloc[:,:5]
original.drop(columns=['original_make_sense_wei'],inplace=True)
original
```

```python
zero_count=0
for i in range(len(original)):
    scores=original.iloc[i,1:]
    if len(scores[scores==0])>=2:
        zero_count+=1
        
print(zero_count)

original['avg_score_o']=0
for i in range(len(original)):
    scores=original.iloc[i,1:]
    if len(scores[scores==0])<2:
        original.iloc[i,4]=round(sum(original.iloc[i,1:4])/3,2)
print(original)

print(sum(original.iloc[:,4])/len(original.iloc[:,4]))
```

```python
questions=np.arange(1,len(original)+1)
#questions
zero_lst=[]
one_lst=[]
two_lst=[]
three_lst=[]
four_lst=[]
five_lst=[]
for i in range(len(original)):
    scores=original.iloc[i,1:4]
    zero_lst.append(len(scores[scores==0]))
    one_lst.append(len(scores[scores==1]))
    two_lst.append(len(scores[scores==2]))
    three_lst.append(len(scores[scores==3]))
    four_lst.append(len(scores[scores==4]))
    five_lst.append(len(scores[scores==5]))
print(len(zero_lst))
inter_coder=pd.DataFrame(list(zip(questions,zero_lst,one_lst,two_lst,three_lst,four_lst,five_lst)),columns = ['q_num','0', '1','2','3','4','5'])
inter_coder.set_index('q_num',inplace=True)
inter_coder
```
